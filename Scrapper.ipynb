```python
# import libraries
import requests
from bs4 import BeautifulSoup

# specify url
url = 'http://prajnaraipn.blogspot.com/'

# request html
page = requests.get(url)

# Parse html using BeautifulSoup, you can use a different parser like lxml if present
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# find searches the given tag (div) with given class attribute and returns the first match it finds
win_div = soup.findAll('div', class_ = 'post-body entry-content')
```

```python
win_div
```

```python
for row in soup.find_all('div',attrs={"class" : "post-body entry-content"}):
    print(row.text)
```
